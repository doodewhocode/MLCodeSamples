# Model prediction code. 

```python
%%writefile prediction.py
#The above option writes all the below lines of python code into prediction.py file. Later we will execute this file from streamlit. 
import cv2
import numpy as np
import streamlit as st
import tensorflow as tf
from tensorflow.keras.preprocessing import image
from tensorflow.keras.preprocessing.image import ImageDataGenerator
import urllib.request

#Download the Saved Model and load the model
urllib.request.urlretrieve("https://raw.githubusercontent.com/venkatareddykonasani/Datasets/master/malaria_detection_models/malaria_detection_30epochs_v2.h5", "malaria_detection_30epochs_v2.h5")
model = tf.keras.models.load_model("malaria_detection_30epochs_v2.h5")

### load file - Cofe for reading the image files uploaded by the user on the web-app
uploaded_file = st.file_uploader("Choose a image file", type="png")

if uploaded_file is not None:
    # Read the imagefile and reshape it - This is pre-processing of the image. Like size and shape. 
    file_bytes = np.asarray(bytearray(uploaded_file.read()), dtype=np.uint8)
    opencv_image = cv2.imdecode(file_bytes, 1)
    opencv_image = cv2.cvtColor(opencv_image, cv2.COLOR_BGR2RGB)
    resized = cv2.resize(opencv_image,(128,128))

    # Displaying the image on the web-app
    st.image(opencv_image, channels="RGB")

    #Final image for prediction.
    img_reshape = resized[np.newaxis,...]
    
    #Predictions will be in numbers, display the respective classes
    map_dict = {0: 'Parasitized', 1: 'Un-infected'}

    #The predict button on webapp
    Genrate_pred = st.button("Predict using the model")

    #Final prediction using the model 
    if Genrate_pred:
        prediction=int(model.predict(img_reshape)) # Model prediction step
        st.title("Predicted Label => {}".format(map_dict[prediction])) #Display the prediction
```

```python
# Run the above python code in streamlit. 
# nohup command will ignore the hang-up signal and the code will run the back-ground 
!nohup streamlit run --server.port 80 prediction.py &
```
